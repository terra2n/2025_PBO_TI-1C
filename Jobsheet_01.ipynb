# Tipe Dasar Data


```python
# 1. Number (int, float, complex)

# Integer (int) : bilangan bulat tanpa desimal.
angka_int = 10
print("angka_int =", angka_int, "->", type(angka_int))
display(Markdown("**Integer (int):** Tipe data yang merepresentasikan bilangan bulat, seperti 10, -5, 0, dan sebagainya. Tidak memiliki bagian desimal."))

# Float : bilangan yang memiliki desimal.
angka_float = 3.14
print("angka_float =", angka_float, "->", type(angka_float))
display(Markdown("**Float:** Tipe data yang merepresentasikan bilangan yang memiliki desimal atau pecahan, seperti 3.14, -2.5, 0.0, dan sebagainya."))

# Complex : bilangan yang memiliki bagian real dan imajiner.
angka_complex = 2 + 3j
print("angka_complex =", angka_complex, "->", type(angka_complex))
display(Markdown("**Complex:** Tipe data yang merepresentasikan bilangan kompleks, yang memiliki bagian real dan imajiner. Bilangan imajiner di Python dilambangkan dengan huruf `j`."))
```

```python
# 2. Operasi Aritmetika
a = 10
b = 3

# Penjelasan tentang operasi aritmetika
display(Markdown("""
**Operasi Aritmetika di Python:**

Operasi aritmetika digunakan untuk melakukan perhitungan matematika pada bilangan. Python mendukung berbagai operasi aritmetika, seperti penjumlahan, pengurangan, perkalian, pembagian, pembagian bulat, modulus (sisa bagi), dan pemangkatan.

**Operator Aritmetika:**

| Operator | Deskripsi |
|---|---|
| `+` | Penjumlahan |
| `-` | Pengurangan |
| `*` | Perkalian |
| `/` | Pembagian (hasil float) |
| `//` | Pembagian Bulat (hasil integer) |
| `%` | Modulus (sisa bagi) |
| `**` | Pemangkatan |
"""))

penjumlahan = a + b      # Tambah
pengurangan = a - b      # Kurang
perkalian = a * b        # Kali
pembagian = a / b        # Bagi (hasil float)
pembagian_bulat = a // b # Bagi (bulat)
modulus = a % b          # Sisa bagi
pangkat = a ** b         # Pemangkatan

print("\nOPERASI ARITMETIKA")
print("a =", a, ", b =", b)
print("Penjumlahan =", penjumlahan)
print("Pengurangan =", pengurangan)
print("Perkalian =", perkalian)
print("Pembagian =", pembagian)
print("Pembagian Bulat =", pembagian_bulat)
print("Modulus =", modulus)
print("Pangkat =", pangkat)
```

```python
# 3. Operasi Perbandingan
# Menghasilkan nilai Boolean (True/False)
a = 10  # Asumsi nilai a dan b dari kode sebelumnya
b = 3

# Penjelasan tentang operasi perbandingan
display(Markdown("""
**Operasi Perbandingan di Python:**

Operasi perbandingan digunakan untuk membandingkan dua nilai. Hasil dari operasi perbandingan adalah nilai Boolean (`True` atau `False`).

**Operator Perbandingan:**

| Operator | Deskripsi |
|---|---|
| `>` | Lebih besar dari |
| `<` | Kurang dari |
| `==` | Sama dengan |
| `!=` | Tidak sama dengan |
| `>=` | Lebih besar dari atau sama dengan |
| `<=` | Kurang dari atau sama dengan |
"""))

lebih_besar = a > b
kurang_dari = a < b
sama_dengan = a == b
tidak_sama = a != b
lebih_besar_sama = a >= b
kurang_sama = a <= b

print("\nOPERASI PERBANDINGAN")
print("a > b =", lebih_besar)
print("a < b =", kurang_dari)
print("a == b =", sama_dengan)
print("a != b =", tidak_sama)
print("a >= b =", lebih_besar_sama)
print("a <= b =", kurang_sama)  # Memperbaiki kesalahan: mbprint menjadi print
```

```python
# 4. Operasi Logika
# and, or, not
x = True
y = False

# Penjelasan tentang operasi logika
display(Markdown("""
**Operasi Logika di Python:**

Operasi logika digunakan untuk menggabungkan atau memodifikasi nilai Boolean (`True` atau `False`). Python memiliki tiga operator logika utama: `and`, `or`, dan `not`.

**Tabel Kebenaran:**

| Operator | Deskripsi | Contoh (x = True, y = False) | Hasil |
|---|---|---|---|
| `and` | Bernilai True jika kedua operand bernilai True | `x and y` | False |
| `or` | Bernilai True jika setidaknya satu operand bernilai True | `x or y` | True |
| `not` | Membalik nilai Boolean dari operand | `not x` | False |
"""))

logika_and = x and y
logika_or = x or y
logika_not_x = not x

print("\nOPERASI LOGIKA")
print("x =", x, ", y =", y)
print("x and y =", logika_and)
print("x or y =", logika_or)
print("not x =", logika_not_x)
```

```python
# 5. Contoh penggunaan di dalam percabangan
a = 10  # Asumsi nilai a dan b dari kode sebelumnya
b = 3

display(Markdown("""
**Contoh penggunaan operasi logika dalam percabangan `if-else`:**

Kode ini menunjukkan contoh penggunaan operasi logika `and` dalam percabangan `if-else`.

**Cara kerja:**

1. **Kondisi:**  `if a > b and b > 0:` memeriksa dua kondisi:
   - `a > b`: Apakah nilai `a` lebih besar dari `b`?
   - `b > 0`: Apakah nilai `b` lebih besar dari 0?
2. **`and`:** Operator `and` mengharuskan kedua kondisi bernilai `True` agar keseluruhan kondisi bernilai `True`.
3. **Eksekusi blok `if`:** Jika kedua kondisi terpenuhi (bernilai `True`), maka blok kode di dalam `if` akan dieksekusi, yaitu mencetak pesan "Kondisi terpenuhi: a lebih besar dari b, dan b masih positif.".
4. **Eksekusi blok `else`:** Jika salah satu atau kedua kondisi tidak terpenuhi (bernilai `False`), maka blok kode di dalam `else` akan dieksekusi, yaitu mencetak pesan "Kondisi tidak terpenuhi atau b <= 0.".
"""))

if a > b and b > 0:
  print("\nKondisi terpenuhi: a lebih besar dari b, dan b masih positif.")
else:
  print("\nKondisi tidak terpenuhi atau b <= 0.")
```

```python
# 4. BREAK pada Loop
print("4) BREAK di dalam loop")
for i in range(10):
    if i == 3:
        print("Loop dihentikan pada i =", i)
        break  # mengakhiri loop saat i = 3
    print("i =", i)
# Keyword break langsung menghentikan keseluruhan perulangan
print()
```

```python
# 5. CONTINUE pada Loop
print("5) CONTINUE di dalam loop")
for i in range(5):
    if i == 2:
        print("Lewati i =", i, "dengan continue")
        continue  # melewati iterasi saat ini dan lanjut ke iterasi berikutnya
    print("i =", i)
# Saat i = 2, baris print("i =", i) tidak akan dieksekusi
print()
```

```python
# 6. NESTED Loop (Loop Bersarang)
print("6) NESTED loop")
for i in range(3):            # Loop luar (outer loop)
    for j in range(2):        # Loop dalam (inner loop)
        print(f"i={i}, j={j}")
# Pada setiap iterasi i, loop j akan berjalan dari 0 sampai 1
print()
```

# Progam Body Mass Index

```python
1import os  # Mengimpor modul os untuk membersihkan layar

def bmi_calculator():
    """Fungsi untuk menghitung dan menampilkan hasil BMI."""
    while True:  # Loop utama, program berjalan terus hingga pengguna memilih keluar
        os.system("cls" if os.name == "nt" else "clear")  # Membersihkan layar (Windows/Unix)
        print("===== Kalkulator Indeks Massa Tubuh (BMI) =====")

        while True:  # Loop untuk input berat dan tinggi, memastikan input valid
            try:
                berat = float(input("Masukkan berat badan Anda (kg): "))  # Input berat (float)
                tinggi_cm = float(input("Masukkan tinggi badan Anda (cm): "))  # Input tinggi (float)
                tinggi_m = tinggi_cm / 100  # Konversi tinggi ke meter
                break  # Keluar dari loop input jika input valid
            except ValueError:  # Menangani error jika input bukan angka
                print("Input tidak valid! Harap masukkan angka.")

        bmi = berat / (tinggi_m ** 2)  # Menghitung BMI
        print(f"\nBMI Anda: {bmi:.2f}")  # Menampilkan BMI dengan 2 angka di belakang koma

        # Menentukan kategori BMI
        if bmi < 18.5:
            print("Kategori: Berat badan kurang (Underweight)")
        elif 18.5 <= bmi < 25:
            print("Kategori: Normal")
        elif 25 <= bmi < 30:
            print("Kategori: Berat badan berlebih (Overweight)")
        else:
            print("Kategori: Obesitas")

        while True:  # Loop untuk input konfirmasi keluar, memastikan input valid
            keluar = input("Apakah Anda ingin keluar? (Y/N): ").strip().lower()  # Input konfirmasi (lowercase)
            if keluar in ("y", "n"):  # Memeriksa apakah input valid ("y" atau "n")
                break  # Keluar dari loop konfirmasi jika input valid
            else:
                print("Input tidak valid! Harap masukkan Y atau N.")

        if keluar == "y":  # Keluar dari loop utama jika pengguna memilih "y"
            break

    print("==== Terima kasih telah menggunakan kalkulator BMI! ====")  # Pesan penutup

bmi_calculator()  # Memanggil fungsi untuk menjalankan program
```

# Progam Pengecekan Bilangan


```python
import os

def isPrime(angka):
    if angka <= 1:
        return False
    for i in range(2, int(angka**0.5) + 1):
        if angka % i == 0:
            return False
    return True

while True:
    os.system("cls" if os.name == "nt" else "clear")  # Membersihkan layar
    print("===== Program Bilangan =====")
    while True:
        try:
            angka = int(input("Silahkan Masukkan Bilangan Bulat Yang Ingin Diperiksa : "))
            break
        except ValueError:
            print("Input Tidak Valid, Silahkan Masukkan Bilangan Bulat")
    # Memeriksa ganjil genap
    if angka % 2 == 0:
        print(f"\nAngka {angka} adalah bilangan Genap")
    else:
        print(f"\nAngka {angka} adalah bilangan Ganjil")
    # Memeriksa Bilangan Prima
    if isPrime(angka):
        print(f"Angka {angka} adalah bilangan Prima")
    else:
        print(f"Angka {angka} BUKAN bilangan Prima")

    # Meminta input untuk melanjutkan atau keluar
    while True:
        keluar = input("Apakah Anda Ingin Keluar (Y/N) : ").strip().upper()
        if keluar in ("Y", "N"):
            break
        else:
            print("Input Tidak Valid! Harap masukkan Y atau N.")

    if keluar == "Y":
        break

print("===== Terimakasih =====")
```

Penjelasan Tambahan:

import os: Mengimpor modul os yang digunakan untuk membersihkan layar.
def isPrime(angka):: Mendefinisikan fungsi isPrime untuk memeriksa apakah suatu bilangan adalah bilangan prima.
Loop while True: Digunakan untuk menjalankan program terus menerus hingga pengguna memilih untuk keluar.
try-except: Digunakan untuk menangani error jika input bukan bilangan bulat.
Pemeriksaan Genap/Ganjil: Menggunakan operator modulus (%) untuk memeriksa apakah bilangan habis dibagi 2.
Pemeriksaan Prima: Memanggil fungsi isPrime untuk memeriksa apakah bilangan prima.
Input Keluar: Meminta input dari pengguna untuk melanjutkan atau keluar.
Pesan Penutup: Menampilkan pesan "===== Terimakasih =====" setelah keluar dari program.
Note: Menambahkan penjelasan singkat tentang program di akhir kode menggunakan triple quotes ("""Docstring""").
